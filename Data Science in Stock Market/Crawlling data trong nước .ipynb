```python
import requests
import os
import pandas as pd
import json
```

```python
url=['https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=ACB&from=1642143906&to=1676271966',
     'https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=BCM&from=1642145145&to=1676273205',
     'https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=BID&from=1642146331&to=1676274391',
     'https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=CTG&from=1642145068&to=1676273128',
     'https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=BVH&from=1642145559&to=1676273619',
     'https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=FPT&from=1642145009&to=1676273069',
     'https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=GAS&from=1642145486&to=1676273546',
     'https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=GVR&from=1642144937&to=1676272997',
     'https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=MBB&from=1642144859&to=1676272920',
     'https://iboard.ssi.com.vn/dchart/api/history?resolution=D&symbol=VCB&from=1642144744&to=1676272804']
```

```python
data= {}
k=0
for i in url:
    request= requests.get(i)
    if request.status_code == 200:
        print("Truy cap thanh cong",k)
    else:
        print(request.status_code)
    dt= request.json()
    data[k]=pd.DataFrame(dt)
    k+=1
```

```python
for i in data:
    data[i]['t']= pd.to_datetime(data[i]['t'], unit='s')
```

```python
name_cp=['ACB','BCM','BID','CTG','BVH','FPT','GAS','GVR','MBB','VCB']
k=0
for i in data:
    data[i]=data[i].replace({'s' : {'ok':name_cp[k]}}, regex=True)
    print(name_cp[k])
    k=k+1
```

```python
for i in data:
    print(data[i])
```

```python
df=data[0]
for i in range(1,10):
    df=pd.concat([df,data[i]], axis=0,ignore_index=1,sort=1)
df['s'].value_counts()
```

```python
new_columns = ["Close", "High", "Low",'Open','company_name',"Date",'Volume']
df = df.rename(columns=dict(zip(df.columns, new_columns)))
df.tail()
```
